```python
dailylife = pd.read_csv("Datasets/LGBT_Survey_DailyLife.csv")
dailylife.head()
```

```python
#analyze by question, group by question given to participants. 
#maybe do select countries?

#Find out countries that were listed
countries = dailylife.groupby("CountryCode").count()
countries

#Pick 3-5 countries, one from each region
#Pick Spain, UK, Germany, Denmark, Romania
```

```python
#Find out questions listed
questions = pd.DataFrame(dailylife.loc[:, ["question_code", "question_label"]]);
#questions.groupby("question_label").value_count()
questions = questions["question_label"].value_counts()
questions
#Select specific questions since there are so many of them to choose
```

```python
#Measure by question and pick by country

#Select by country
nations_chosen = ["Spain", "United Kingdom", "Romania"]
picked_nations = dailylife[dailylife.CountryCode.isin(nations_chosen)]
picked_nations
```

```python
#Select social-based questions that focuses on daily life aspects
questions = ["Have you been open about you being L, G, B or T?", "In your opinion, how many people know that you are L, G, B or T? Friends?",
            "To how many people among the following groups are you open about yourself being L, G, B or T? Family members (other than your partner(s))? "]
picked_questions = dailylife[dailylife.question_label.isin(questions)]
picked_questions
```

```python
uk = picked_questions[picked_questions["CountryCode"] == "United Kingdom"]
uk
```

```python
uk_survey = pd.DataFrame(uk.groupby(['subset','answer']).sum()['percentage'].unstack())
uk_survey
```

```python
uk.groupby(['subset','answer']).count()['percentage'].unstack(fill_value=0).plot.bar(figsize = (30,10), width = 1)
plt.title("Survey Responses in UK")
plt.grid()
```
